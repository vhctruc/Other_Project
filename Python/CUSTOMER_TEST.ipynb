```python
# Approximate radius of earth in km

xlat1 = radians(16.100858)
xlon1 = radians(108.249777)
xlat2 = radians(16.068066)
xlon2 = radians(108.214012)

def cal_radius(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    R = 6371.0

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
#     c = 2 * atan2(sqrt(a), sqrt(1 - a))
    c = 2 * asin(sqrt(a))
    distance = R * c
    return distance
#     print("Result: ", distance)

# cal_radius(xlat1,xlon1,xlat2,xlon2)
# cal_radius(lat1, lon1, lat2, lon2)
```
